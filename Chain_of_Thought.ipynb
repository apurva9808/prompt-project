# **Chain of Thought (CoT) Exercise**

A Chain of Thought (CoT) instruction is really only needed when the question is tricky and the AI requires extra details to grasp the issue. CoT is a method that helps us explain the reasoning behind an answer. We use this approach alongside few-shot learning to overcome some of the restrictions that come with few-shot learning.

# **Without COT**

```python
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def chat_with_felix():
    """Chat with Felix, the chatbot."""
    print("Felix: Hi there. I am Felix, the chatbot. Let's play a game.")

    while True:
        message = input("You: ")
        if message.lower() in ['exit', 'quit']:
            print("Felix: Goodbye!")
            break

        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a smart and helpful assistant."},
                {"role": "user", "content": f"{message}"}
            ],
            temperature=1,
            max_tokens=256,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0
        )

        try:
            print("Felix:", response.choices[0].message.content)
        except:
            print("Felix: Sorry, a problem occurred. Please try again later.")

if __name__ == '__main__':
    chat_with_felix()
```

# **CoT Implementation**

```python
import os
from openai import OpenAI

client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def chat_with_felix():
    """Chat with Felix, the chatbot."""
    print("Felix: Hi there. I am Felix, the chatbot. Let's play a game.")

    while True:
        message = input("You: ")
        if message.lower() in ['exit', 'quit']:
            print("Felix: Goodbye!")
            break

        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a smart and helpful assistant."},
                {"role": "user", "content": "When James was 2 years old, his sister was 2*2 years old. James is now 30 years old. How old is his sister?"},
                {"role": "assistant", "content": "When James was 2 years old, his sister was 4 years old. There's 2 year differences between them and the sister is older than James. When James is 30 years old, his sister should be 32 years old."},
                {"role": "user", "content": "When Harry was 25 years old, his sister was 29 years old. Harry is now 30 years old. How old is his sister?"},
                {"role": "assistant", "content": "When Harry was 25 years old, his sister was 29 years old. There's 4 year differences between them and the sister is older than Harry. When Harry is 30 years old, his sister should be 34 years old."},
                {"role": "user", "content": f"{message}"}
            ],
            temperature=1,
            max_tokens=256,
            top_p=1,
            frequency_penalty=0,
            presence_penalty=0
        )

        try:
            print("Felix:", response.choices[0].message.content)
        except:
            print("Felix: Sorry, a problem occurred. Please try again later.")

if __name__ == '__main__':
    chat_with_felix()
```

Certain topics, like symbolic and arithmetic reasoning, are challenging for AI models like GPT. These topics often require a technique called Chain-of-Thought (CoT) to help the model make sense of the problem. This is because the model doesn't truly understand the problem itself; it tries to predict the answer based on patterns it has learned, rather than actual logic.

In our case, using CoT lets the model dissect the problem into simpler steps. It's like giving the model a clear map that shows how to navigate the problem, helping it piece together the reasoning needed to find the answer.

# **Zero-Shot CoT Prompting**

There might be times when you can't or don't want to use examples for a Chain-of-Thought instruction. In those situations, zero-shot CoT prompting could be a good alternative to try.
